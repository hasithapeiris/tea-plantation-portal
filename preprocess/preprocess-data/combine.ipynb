```python
import pandas as pd

# Load preprocessed datasets
export_data = pd.read_csv("processed_tea_export_data.csv")
production_data = pd.read_csv("processed_tea_production_data.csv")
exchange_rate_data = pd.read_csv("processed_exchange_rate_data.csv")
```

```python
# Merge datasets on 'Year-Month'
combined_data = pd.merge(export_data, production_data, on="Year-Month", how="inner")
combined_data = pd.merge(combined_data, exchange_rate_data, on="Year-Month", how="inner")
```

```python
# Calculate FEE (Foreign Exchange Earnings)
combined_data['FEE_(USD)'] = (
    combined_data['Export_Quantity_(Kg)'] * combined_data['Export_Price_(Rs.)'] / combined_data['Avg_Exchange_Rate']
)
```

```python
# Rearrange columns for clarity
combined_data = combined_data[[
    'Year-Month', 'Export_Quantity_(Kg)', 'Export_Price_(Rs.)', 
    'High_Grown_Quantity_(Kg)', 'Medium_Grown_Quantity_(Kg)', 'Low_Grown_Quantity_(Kg)', 
    'Avg_Exchange_Rate', 'FEE_(USD)'
]]
```

```python
combined_data.to_csv("combined_dataset.csv", index=False)
```
