```python
import pandas as pd
```

```python
# Load datasets
lgp_df = pd.read_csv("lgp_forecast.csv")
mgp_df = pd.read_csv("mgp_forecast.csv")
hgp_df = pd.read_csv("hgp_forecast.csv")
```

```python
# Merge on 'Month' column
merged_df = lgp_df.merge(mgp_df, on='month', how='outer').merge(hgp_df, on='month', how='outer')

# Set 'Month' as the index
merged_df.set_index('month', inplace=True)
```

```python
print(merged_df)
```

```python
merged_df.to_csv('merged_elevations.csv')
```
