```python
import pandas as pd

forex_df = pd.read_csv("forex_dataset.csv")
```

```python
forex_df.head()
```

```python
# Convert 'Year-Month' to a datetime object
forex_df['Year-Month'] = pd.to_datetime(forex_df['Year-Month'])
```

```python
forex_df = forex_df.sort_values('Year-Month')
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the Foreign Exchange Earnings (FEE) over time
plt.figure(figsize=(12, 6))
plt.plot(forex_df['Year-Month'], forex_df['FEE_(USD)'], marker='o', linestyle='-', color='g', label='FEE (USD)')
plt.title('Foreign Exchange Earnings Over Time', fontsize=12)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('FEE (USD)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
#plt.tight_layout()
```

```python
models = ['Linear Regression', 'Random Forest', 'ARIMA']
r2_scores = [0.96, 0.72, 0.88]

plt.figure(figsize=(10, 6))
plt.bar(models, r2_scores, color=['blue', 'green', 'orange'])
plt.title('Model Performance (R Score)', fontsize=16)
plt.ylabel('R Score', fontsize=12)
plt.ylim(0, 1) 
plt.grid(axis='y', alpha=0.5)
plt.show()
```
